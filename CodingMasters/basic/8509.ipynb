```python
# 커트라인 정하기

# 입력 받기
# N과 K를 입력으로 받음
N, K = map(int, input().split())

# 각 학생의 점수를 리스트로 입력받음
score = list(map(int, input().split()))

# 임계값 후보를 저장할 리스트 초기화
cut_line = []
for s in score:
    # 중복되지 않는 점수만 추가
    if s not in cut_line:
        cut_line.append(s)

# 점수 중 최솟값을 찾음
minimum = min(score)

# 각 임계값 후보마다 학생 수를 카운트하고, 그 수가 K 이하인 경우를 임계값으로 선정
cnt = []
result_cut_line = []
for cut in cut_line:
    count = 0
    for s in score:
        if s >= cut:
            count += 1
        else:
            # 가장 낮은 점수일 때
            if score.index(s) == 0:
                if score[1] >= cut:
                    count += 1
            # 가장 높은 점수일 때
            elif score.index(s) == len(score) - 1:
                if score[-2] >= cut:
                    count += 1
            # 그 외의 경우
            else:
                if score[score.index(s) - 1] >= cut or score[score.index(s) + 1] >= cut:
                    count += 1

    # 카운트가 K 이하일 때 임계값과 카운트 수를 저장
    if count <= K:
        cnt.append(count)
        result_cut_line.append(cut)

# 최대값 및 해당 인덱스 찾기
if cnt:  # cnt 리스트가 비어있지 않은 경우
    max_val = max(cnt)  
    max_index = cnt.index(max_val)  
else:  # cnt 리스트가 비어있는 경우
    max_val = 0  
    max_index = 0  

# 최대값이 여러 개인 경우, 가장 큰 임계값을 가진 인덱스 찾기
for i in range(len(cnt)):
    if max_val == cnt[i] and result_cut_line[i] > result_cut_line[max_index]:
        max_val = cnt[i]  # 최대값 업데이트
        max_index = i  # 해당 인덱스 업데이트

# 결과 출력
if max_val == 0:  # 만약 최대값이 0이면
    print(minimum - 1)  # 최솟값에서 1을 뺀 값을 출력
else:
    print(result_cut_line[max_index])  # 그렇지 않으면 최대값에 해당하는 임계값을 출력
```
